```python
from cad_service.cad_machining_visualization_pyvista import CADMachiningVisualizationPyVista
from cad_service.kernel.cad_loader import CADLoader
import os
from pathlib import Path
import shutil
from llm_utils.communication_utils.models.view_orientation import ViewOrientation



resources_dir = Path(os.path.abspath("")).parent.parent / "resources"
cad_ds_dir = resources_dir / "cad-models"
# if abc_out_dir.exists():
#     shutil.rmtree(abc_out_dir)

for ds_dir in cad_ds_dir.iterdir():
    if not ds_dir.is_dir():
        continue

    raw_ds_dir = ds_dir / "raw"
    processed_ds_dir = ds_dir / "processed"
    processed_ds_dir.mkdir(exist_ok=True)
    for sample in raw_ds_dir.iterdir():
        idx = sample.stem.split("_")[0]
        print(idx)

        abc_sample_dir = processed_ds_dir / idx

        if abc_sample_dir.exists():
            continue
        abc_sample_dir.mkdir()

        shutil.copyfile(sample, abc_sample_dir / "cad.step")

        shape = CADLoader().read_shape(sample)
        img = CADMachiningVisualizationPyVista().screenshot_shape(shape, orientation=ViewOrientation.FRONT)
        img.save(abc_sample_dir / "screenshot.png")
        img.show()

        text = input("")

        with open(abc_sample_dir / 'input.txt', 'w') as f:
            f.write(text)
```
